## Q4. Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curv

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=7,random_state=42)
km.fit_predict(x)
```

```python
wcss=[]
for i in range(2,7):
    kmeans=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,tol=0.0001,verbose=0,random_state=None,copy_x=True,algorithm='auto')
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,7),wcss)
```

## From the above graph I found that optimal number of cluster using elbow method is 3

### Q5. Find the optimum value of k for k-means clustering using the silhouette score method and specify the number of 3observations in each cluster using a bar plot

```python
import sklearn.metrics as metrics
for i in range(2,7):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=200)
    kmeans.fit(df)
    labels=kmeans.labels_
    print("Silhouette  score for k(clusters)="+str(i)+" is "+str(metrics.silhouette_score(df,labels,metric="euclidean",sample_size=1000,random_state=200)))
```

## From the above observation using silhouette score method optimal number of cluster is found to be 2
